```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('top_1000_pe_imports.csv')
```

```python
df
```

```python
x = df.drop(['hash','malware'],axis=1) 
```

```python
y = df['malware']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101) #훈련 셋, 테스트 셋으로 나누기
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 5, min_samples_split = 20) #decision tree로 모델링
```

```python
dtree.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score #교차검증 적용

score=cross_val_score(dtree, x_train, y_train, cv=5) 

print('cross validation score : {:.3f}'.format(score.mean()))
```

```python
predicties = dtree.predict(x_test) #테스트 셋에 대한 예측값
```

```python
predicties
```

```python
i = 1
mal =0
good =0
for each in predicties : #예측값이
    if each == 1: #1이라면 malware
        mal +=1
        print("NO.%d is Malware." %(i))
    else: #0이라면 goodware
        good +=1
        print("NO.%d is Goodware." %(i))
    i+=1
```

```python
print('total number:',i)
print('malware number :',mal)
print('goodware number : ',good)
```

```python
a = tree.plot_tree(dtree, feature_names=df.columns, filled = True, class_names=['malware', 'goodware']) #폰트 사이즈 조절하여 상세 내역 확인
```

```python
import sklearn.metrics as mt #모델에 대한 정확도
```

```python
accuracy = mt.accuracy_score(y_test, predicties)
recall = mt.recall_score(y_test, predicties)
precision = mt.precision_score(y_test, predicties)
f1_score = mt.f1_score(y_test, predicties)

print("Accuracy: {:.3f}" .format(accuracy))
print("Recall: {:.3f}" .format(recall))
print("Precision: {:.3f}" .format(precision))
print("F1_score: {:.3f}".format(f1_score))
```
